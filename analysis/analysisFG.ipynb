```python
import json
import numpy as np

from graphviz import Digraph

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = {}
with open('../data/FG.json', 'r') as f:
    data = json.load(f)
```

```python
maxFitness = np.zeros((len(data.keys()), 100000))
avgFitness = np.zeros((len(data.keys()), 100000))

for iter in data.keys():
    maxFitness[int(iter), :] = np.array(data[iter]['maxFitness'])
    avgFitness[int(iter), :] = np.array(data[iter]['avgFitness'])
```

```python
iter = 0
max = maxFitness[iter, :]
avg = avgFitness[iter, :]


plt.figure().set_figwidth(15)
plt.plot(max, label='Max Fitness')
plt.plot(avg, label='Average Fitness')
plt.ylim(0, 1.05);
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.legend();
```
